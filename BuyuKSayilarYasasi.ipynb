Yazı tura atışında atış sayısı sonsuza yaklaştığında oran  %50 ye doğru geliyor.

```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 

liste=[] 

for i in np.arange(0,10) : 
    denemesayisi=i**3 
    yt=np.random.randint(0,2,size=denemesayisi) 
    yolasılık=np.mean(yt) 
    liste.append(yolasılık) 

sns.lineplot (data=liste, linewidth=2) 
plt.xlabel("Deneme Sayısı") 
plt.ylabel("Ortalama") 
plt.ylim (0,1) 
plt.axhline (0.5, linestyle="--", linewidth=1.5, color="red")
plt.show()
```
