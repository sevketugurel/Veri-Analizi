```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
np.random.seed(0) # seed yapısı oluşturulan verinin sürekli değişmesini engeller
np.random.seed (0) 
veri=np.random.normal (35,2,40000) #  ortalaması 35 standard sapması 2 olan 4000 tane  veri oluşturuyoruz
veriZ = ((veri-np.mean(veri))/np.std(veri)) # formül hali ile verinin zscore'u
veriZ2 = stats.zscore(veri) # stats ile zscore

sns.displot (veriZ, kde=True) 
plt.title("Veri Dağılım Grafiği",fontsize=15, loc="right",c="red") 
plt.xlabel("Veriler", fontsize=10,c="red")
plt.ylabel("Frekans", fontsize=10,c="red") 
plt.xlim(-3,+3) # std göstermek için ve sınırları belirledik yani -1 ve +1 bunun sayesinde gözüküyor 
plt.axvline(x=np.mean(veriZ), linestyle="--", linewidth=1.5, label="Ortalama",c="red")
plt.axvline(x=np.mean(veriZ)-np.std(veriZ),linestyle="--",linewidth=1.5, c="black")
plt.axvline(x=np.mean(veriZ)+np.std(veriZ),linestyle="--",linewidth=1.5, label="1 Standart Sapma", c="black")
plt.axvline(x=np.mean(veriZ)-2*np.std(veriZ),linestyle="--",linewidth=1.5, label="1 Standart Sapma", c="blue")
plt.axvline(x=np.mean(veriZ)+2*np.std(veriZ),linestyle="--",linewidth=1.5, label="1 Standart Sapma", c="blue")
plt.legend() # açıklamaları gözterebilmek için
plt.show()

```
