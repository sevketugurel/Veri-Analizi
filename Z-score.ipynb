```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
np.random.seed(0) # seed yaps oluturulan verinin srekli deimesini engeller
np.random.seed (0) 
veri=np.random.normal (35,2,40000) #  ortalamas 35 standard sapmas 2 olan 4000 tane  veri oluturuyoruz
veriZ = ((veri-np.mean(veri))/np.std(veri)) # forml hali ile verinin zscore'u
veriZ2 = stats.zscore(veri) # stats ile zscore

sns.displot (veriZ, kde=True) 
plt.title("Veri Dalm Grafii",fontsize=15, loc="right",c="red") 
plt.xlabel("Veriler", fontsize=10,c="red")
plt.ylabel("Frekans", fontsize=10,c="red") 
plt.xlim(-3,+3) # std gstermek iin ve snrlar belirledik yani -1 ve +1 bunun sayesinde gzkyor 
plt.axvline(x=np.mean(veriZ), linestyle="--", linewidth=1.5, label="Ortalama",c="red")
plt.axvline(x=np.mean(veriZ)-np.std(veriZ),linestyle="--",linewidth=1.5, c="black")
plt.axvline(x=np.mean(veriZ)+np.std(veriZ),linestyle="--",linewidth=1.5, label="1 Standart Sapma", c="black")
plt.axvline(x=np.mean(veriZ)-2*np.std(veriZ),linestyle="--",linewidth=1.5, label="1 Standart Sapma", c="blue")
plt.axvline(x=np.mean(veriZ)+2*np.std(veriZ),linestyle="--",linewidth=1.5, label="1 Standart Sapma", c="blue")
plt.legend() # aklamalar gzterebilmek iin
plt.show()

```
